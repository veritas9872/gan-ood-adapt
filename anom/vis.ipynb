```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
f = '../data/transistor/train/good/000.png'
img = np.asarray(Image.open(f).convert('L'))
ftr = np.fft.fft2(img, norm='ortho').real
print(ftr.shape, ftr.min(), ftr.max())
plt.imshow(np.log10(np.absolute(ftr)), cmap='gray')
plt.tick_params(which='both', bottom=False, top=False)
plt.colorbar()
plt.show()
```
